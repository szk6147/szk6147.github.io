```python
'''I chose the actors dataset
These ae the questions:
Q1. Who are the top-grossing actors based on total gross earnings?
Q2. Which actor has the highest average earnings per movie?
Q3. What is the distribution of total gross earnings among actors?
Q4. Is there a correlation between the number of movies an actor has appeared in and their total gross earnings?
Q5. Which actor has starred in the highest-grossing movie?
Q6. How does the average earnings per movie vary among actors?
Q7. Is there any relationship between the average earnings per movie and the number of movies an actor has appeared in?
'''
import pandas as pd
import matplotlib.pyplot as plt

# Load the actors dataset
actors_df = pd.read_csv("actors.csv")

# Check for missing values
print("Checking for missing values:")
print(actors_df.isnull().sum())

# Summary statistics
print("\nSummary statistics:")
print(actors_df.describe())

# Visualize the distribution of total gross earnings
plt.hist(actors_df['Total Gross'], bins=20)
plt.xlabel('Total Gross Earnings')
plt.ylabel('Frequency')
plt.title('Distribution of Total Gross Earnings')
plt.show()

# Calculate the correlation between number of movies and total gross earnings
correlation = actors_df['Number of Movies'].corr(actors_df['Total Gross'])
print(f"\nCorrelation between Number of Movies and Total Gross Earnings: {correlation}")

# Visualize the top-grossing actors
top_grossing_actors = actors_df.sort_values(by='Total Gross', ascending=False).head(10)
plt.barh(top_grossing_actors['Actor'], top_grossing_actors['Total Gross'])
plt.xlabel('Total Gross Earnings')
plt.ylabel('Actor')
plt.title('Top-Grossing Actors')
plt.show()

# Visualize average earnings per movie
plt.barh(actors_df['Actor'], actors_df['Average per Movie'])
plt.xlabel('Average Earnings per Movie')
plt.ylabel('Actor')
plt.title('Average Earnings per Movie for Actors')
plt.show()

# Q1: Who are the top-grossing actors based on total gross earnings?
top_grossing_actors_total_gross = actors_df.sort_values(by='Total Gross', ascending=False).head(5)
print("\nTop-Grossing Actors based on Total Gross Earnings:")
print(top_grossing_actors_total_gross[['Actor', 'Total Gross']])

# Q2: Which actor has the highest average earnings per movie?
highest_avg_earnings_actor = actors_df.loc[actors_df['Average per Movie'].idxmax()]
print("\nActor with the Highest Average Earnings per Movie:")
print(highest_avg_earnings_actor[['Actor', 'Average per Movie']])

# Q3: Distribution of Total Gross Earnings among Actors
plt.hist(actors_df['Total Gross'], bins=20)
plt.xlabel('Total Gross Earnings')
plt.ylabel('Frequency')
plt.title('Distribution of Total Gross Earnings')
plt.show()

# Q4: Correlation between Number of Movies and Total Gross Earnings
correlation = actors_df['Number of Movies'].corr(actors_df['Total Gross'])
print(f"Correlation between Number of Movies and Total Gross Earnings: {correlation}")

# Q5: Which actor has starred in the highest-grossing movie?
highest_grossing_movie_actor = actors_df.loc[actors_df['Gross'].idxmax()]
print("\nActor who has Starred in the Highest-Grossing Movie:")
print(highest_grossing_movie_actor[['Actor', '#1 Movie', 'Gross']])

# Q6: Average Earnings per Movie for Actors
plt.barh(actors_df['Actor'], actors_df['Average per Movie'])
plt.xlabel('Average Earnings per Movie')
plt.ylabel('Actor')
plt.title('Average Earnings per Movie for Actors')
plt.show()


# Q7: Is there any relationship between the average earnings per movie and the number of movies an actor has appeared in?
plt.scatter(actors_df['Number of Movies'], actors_df['Average per Movie'])
plt.xlabel('Number of Movies')
plt.ylabel('Average Earnings per Movie')
plt.title('Relationship between Number of Movies and Average Earnings per Movie')
plt.show()

```
